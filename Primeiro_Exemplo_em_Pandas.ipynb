```python
#aqui criamos um novo dataframe apenas com as colunas que eu quero
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
#exibe a 4a linha
display(ca_df.loc[3])
```

```python
#exibir a 10a linha ate a 20a
display(ca_df.loc[9:19])
```

```python
#criar um dataframe gas_df contendo 
#apenas as 4 colunas (Revenda, municipio, Produto, Valor de Venda)
#somente com combustivel sendo gasolina e exibir na tela
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#exibir qual é o posto e municipio da gasolina mais cara do pais?
display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
#DataFrame.loc[] com múltiplas condições para filtragem
#Quais são os preços, postos que vendem Etanol na minha cidade (divinopolis)
etanol_divinopolis_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'DIVINOPOLIS')]
display(etanol_divinopolis_df)
```
